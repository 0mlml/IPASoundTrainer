```python
from IPython.display import Audio, display
import random
from time import sleep
import os
```

```python
audio_files = os.listdir('/content/content')[1:-1]
print(audio_files)

points = 0
counter = 0
```

```python
path = random.choice(audio_files)
display(Audio(path, autoplay=True))
```

```python
print(path)
```

```python
print("æ" if "a" in path else "ʌ")
print(points, "/", counter, ",", int(points/counter*100), "%")
```

```python
if "a" in path:
  points += 1
counter += 1
```

```python
if "u" in path:
  points += 1
counter += 1
```

```python
print("f" if "f" in path else "th")
print(points, "/", counter, ",", int(points/counter*100), "%")
```
